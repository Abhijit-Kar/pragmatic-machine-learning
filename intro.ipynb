# Python
Parsel Tongue of Nerds

```python
def memoise(func):
    cache = {}
    def wrap(n):
        if n not in cache:
            cache[n] = func(n)      
        return cache[n]
    return wrap
@memoise
def fact(n):
    if n == 1 or n == 0:
        return 1
    return n * fact(n - 1)
for i in range(1, 10):
    print(fact(i))
```

# Matplotlib
Need I say more

```python
import seaborn as sns
sns.set()
sns.set_context('talk')

%matplotlib inline
time = np.arange(0, 20, 0.1);
amplitude = np.cos(time)
```

```python
import matplotlib.pyplot as plt

plt.plot(time, amplitude)
plt.title('Cosine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axhline(y = 0)
plt.show()
```

# Sci Kit Learn
Machine Learning Toolkit of ages

```python
class VizKit():
    @staticmethod
    def show_optimum_k(k_range, scores):
        sns.set_context('talk')

        plt.figure(figsize = (10, 5))
        plt.plot(k_range, scores)
        plt.xlabel('Value of K for KNN')
        plt.ylabel('Testing Accuracy')
        plt.show()

iris_df = pd.read_csv('data/iris.csv', dtype = {'species': 'category'})
X = iris_df.iloc[:, :-1]
y = iris_df.species
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()
grid = GridSearchCV(knn, {'n_neighbors': range(1, 31)}, cv = 10, scoring = 'accuracy', n_jobs = -1)
grid.fit(X, y)

VizKit.show_optimum_k(range(1, 31), grid.cv_results_['mean_test_score'])
```
