```python
import pandas as pd
```

```python
titanic_df = pd.read_csv('../data/titanic.csv')
titanic_df.columns = titanic_df.columns.str.lower()
titanic_df.head(3)
```

# Dummy Variables

```python
titanic_df = pd.get_dummies(titanic_df, columns = ['sex', 'embarked'], drop_first = True)
titanic_df.columns = titanic_df.columns.str.lower()
titanic_df.head(3)
```
