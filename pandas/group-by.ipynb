```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set()
```

```python
drinks_df = pd.read_csv('../data/drinks-by-country.csv')
drinks_df.head(3)
```

```python
drinks_df.groupby('continent').beer_servings.mean()
```

```python
drinks_df.groupby('continent').mean()
```

```python
sns.set_context('talk')

plt.figure(figsize = (15, 5))

drinks_df.groupby('continent').mean().plot(kind = 'bar', figsize = (15, 5))

plt.show()
```

```python
drinks_df.groupby('continent').beer_servings.agg(['count', 'min', 'max', 'mean'])
```

```python
drinks_df.groupby('continent').agg(['count', 'min', 'max', 'mean'])
```

```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
sns.set_context('talk')

with sns.color_palette("Spectral", 12):
    flight_df.groupby('year').mean().plot(kind = 'pie', subplots = True, figsize = (7, 7), legend = False, autopct = '%1.1f%%')

plt.ylabel('')
plt.title('Passengers\nYear by Year')
plt.show()
```
