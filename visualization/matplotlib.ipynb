```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set()
```

```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

# Line Graph

**Need to use Matplotlib for this**

```python
sns.set_context('talk')

plt.figure(figsize = (15, 5))

plt.plot(flight_df.year, flight_df.passengers, label = 'Airbus')
plt.plot(flight_df.year, flight_df.passengers / 2, label = 'Kingfisher', linestyle = 'dashed')

plt.xlabel('Years')
plt.ylabel('Passengers')

plt.title('Passengers\nYear by Year')

plt.legend()
plt.show()
```

# Pie Chart

```python
import numpy as np
```

```python
passengers_by_year = flight_df.groupby('year').agg(np.mean)
```

```python
sns.set_context('talk')

plt.figure(figsize = (7, 7))

with sns.color_palette("Spectral", 12):
    plt.pie(passengers_by_year, labels = flight_df.groupby('year').groups, autopct = '%1.1f%%')

plt.title('Passengers\nYear by Year')
plt.show()
```

# Displaying Images

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
# sns.set_style('dark')
sns.set_context('talk')

# sns.reset_orig()    # If we want Matplotlib styling OR cmap = viridis, RdPu

plt.figure(figsize = (20, 8))

for i, (image, label) in enumerate(zip(digits.data[:10], digits.target[:10])):
    plt.subplot(2, 5, i + 1).axis('off')
    plt.imshow(image.reshape(8, 8), cmap = 'RdPu')
    plt.title('[{}]'.format(label))

plt.show()
```

# Cosine Wave

```python
sns.set_context('talk')

time = np.arange(0, 20, 0.1);
amplitude = np.cos(time)

plt.plot(time, amplitude)

plt.title('Cosine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.axhline(y = 0)

plt.show()
```
