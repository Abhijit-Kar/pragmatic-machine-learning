```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set()
```

```python
pima_df = pd.read_csv('../data/pima.csv', dtype = {'diabetes': int})
pima_df.head(3)
```

```python
drinks_df = pd.read_csv('../data/drinks-by-country.csv')
drinks_df.head(3)
```

```python
flight_df = sns.load_dataset('flights')
flight_df.head(3)
```

```python
ufo_df = pd.read_csv('../data/ufo.csv')
ufo_df.columns = ufo_df.columns.str.lower().str.replace(' ', '_')
ufo_df.time = pd.to_datetime(ufo_df.time)
ufo_df['year'] = ufo_df.time.dt.year
ufo_df.head(3)
```

```python
imdb_df = pd.read_csv('../data/imdb.csv')
imdb_df.head(3)
```

# Bar Chart

```python
sns.set_context('talk')

plt.figure(figsize = (15, 5))

drinks_df.groupby('continent').mean().plot(kind = 'bar', figsize = (15, 5), cmap = 'viridis', rot = 0)

plt.show()
```

```python
sns.set_context('talk')

pd.crosstab(imdb_df.genre, imdb_df.content_rating).plot(kind = 'bar', rot = 45, figsize = (16, 6),
                                                        stacked = True, cmap = 'icefire')

plt.show()
```

# Pie Chart

```python
sns.set_context('talk')

with sns.color_palette("Spectral", 12):
    flight_df.groupby('year').mean().plot(kind = 'pie', subplots = True, figsize = (7, 7), legend = False, autopct = '%1.1f%%')

plt.ylabel('')
plt.title('Passengers\nYear by Year')
plt.show()
```

# Line Chart

```python
sns.set_context('talk')

imdb_df.star_rating.plot(figsize = (10, 4))

plt.show()
```

```python
sns.set_context('talk')

pd.crosstab(imdb_df.genre, imdb_df.content_rating).plot(figsize = (16, 6), rot = 45)

plt.show()
```

# Histogram

```python
sns.set_context('talk')

ufo_df.plot(kind = 'hist', figsize = (10, 4), bins = 20)

plt.show()
```
