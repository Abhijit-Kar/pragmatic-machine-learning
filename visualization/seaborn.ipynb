```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set()
```

```python
pima_df = pd.read_csv('../data/pima.csv')
pima_df.head()
```

```python
iris_df = sns.load_dataset('iris')     # pd.read_csv('../data/iris.csv')
iris_df.species = iris_df.species.astype('category')    # Optimization
iris_df.head()
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

# Dist Plot
* **Helps visualizing Single Variable distribution**

```python
sns.set_context('talk')
plt.figure(figsize = (9, 7))
sns.distplot(pima_df.age)
plt.show()
```

# KDE Plot
* **Kernel Density Plot**
* **Works with 2 Variables**

```python
sns.set_context('talk')
plt.figure(figsize = (9, 7))
sns.kdeplot(pima_df.age, pima_df.glucose_conc, shade = True)
plt.ylabel('Glucose Concentration')
plt.xlabel('Age')
plt.show()
```

# Pair Plot
* **Helps check categorical separation**

```python
sns.set_context('notebook')    # notebook, paper, talk, poster
sns.pairplot(iris_df, hue = 'species')
plt.show()
```

# Strip Plot
* **Bad Idea for large data set**

```python
sns.set_context('poster')
plt.figure(figsize = (15, 10))
sns.stripplot(data = pima_df, y = 'age', x = 'num_preg', jitter = True, hue = 'diabetes', dodge = True)
plt.show()
```

# Swarm Plot
* **Better than strip plot**

```python
sns.set_context('poster')
plt.figure(figsize = (15, 10))
sns.swarmplot(data = pima_df, x = 'num_preg', y = 'age', hue = 'diabetes', dodge = True)
plt.show()
```

# Box Plot

```python
sns.set_context('talk')
plt.figure(figsize = (7, 7))
sns.boxplot(data = iris_df, x = 'species', y = 'sepal_width')
plt.show()
```

# Joint Plot

```python
sns.set_context('talk')
(sns.jointplot(data = iris_df, x = 'sepal_length', y = 'petal_length', size = 8, kind = 'reg')
 .plot_joint(sns.kdeplot))    # Plot Joint added 2d kde plot to the joint plot
plt.show()
```

# Violin Plot

```python
sns.set_context('talk')
plt.figure(figsize = (7, 7))
sns.violinplot(data = pima_df, x = 'diabetes', y = 'num_preg', scale = 'count')
plt.show()
```

# LM Plot
* **Linear Model Plot**

```python
sns.set_context('talk')
sns.lmplot(data = iris_df, x = 'sepal_length', y = 'petal_length', hue = 'species', col = 'species', size = 8)
plt.show()
```

# Point Plot

```python
sns.pointplot(data = pima_df, x = 'diabetes', y = 'age')
plt.show()
```

# Bar Plot

```python
sns.barplot(data = pima_df, x = 'diabetes', y = 'age')
plt.show()
```

# Count Plot

```python
sns.countplot(data = pima_df, x = 'diabetes')
plt.show()
```

# Factor Plot

```python
sns.set_context('notebook')
sns.factorplot(data = titanic_df[titanic_df.embark_town.notnull()],
               x = 'age', y = 'embark_town',
               hue = 'sex', row = 'class',
               orient = 'h', size = 3, aspect = 3.5, palette = 'Set3',
               kind = 'violin', split = True, cut = 0, bw = .2)
plt.show()
```

# Heat Map

```python
sns.set_context('talk')
plt.figure(figsize = (10, 7))
flight_df = flight_df.pivot('month', 'year', 'passengers')
sns.heatmap(flight_df, annot = True, fmt = 'd', cbar = False, cmap = "YlGnBu")
plt.show()
```

```python
sns.set_context('talk')
plt.figure(figsize = (10, 8))
sns.heatmap(pima_df.corr())
plt.show()
```
